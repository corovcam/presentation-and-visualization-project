```python
# All necessary imports
import pandas as pd
import numpy as np

# Start by importing the dataset and filling the missing values with consecutive ones.
data1 = pd.read_csv("PRSA_Data_Aotizhongxin_20130301-20170228.csv")
data1 = data1.ffill()
data1 = data1.bfill()
data2 = pd.read_csv("PRSA_Data_Changping_20130301-20170228.csv")
data2 = data2.ffill()
data2 = data2.bfill()
data3 = pd.read_csv("PRSA_Data_Dingling_20130301-20170228.csv")
data3 = data3.ffill()
data3 = data3.bfill()
data4 = pd.read_csv("PRSA_Data_Dongsi_20130301-20170228.csv")
data4 = data4.ffill()
data4 = data4.bfill()
data5 = pd.read_csv("PRSA_Data_Guanyuan_20130301-20170228.csv")
data5 = data5.ffill()
data5 = data5.bfill()
data6 = pd.read_csv("PRSA_Data_Gucheng_20130301-20170228.csv")
data6 = data6.ffill()
data6 = data6.bfill()
data7 = pd.read_csv("PRSA_Data_Huairou_20130301-20170228.csv")
data7 = data7.ffill()
data7 = data7.bfill()
data8 = pd.read_csv("PRSA_Data_Nongzhanguan_20130301-20170228.csv")
data8 = data8.ffill()
data8 = data8.bfill()
data9 = pd.read_csv("PRSA_Data_Shunyi_20130301-20170228.csv")
data9 = data9.ffill()
data9 = data9.bfill()
data10 = pd.read_csv("PRSA_Data_Tiantan_20130301-20170228.csv")
data10 = data10.ffill()
data10 = data10.bfill()
data11 = pd.read_csv("PRSA_Data_Wanliu_20130301-20170228.csv")
data11 = data11.ffill()
data11 = data11.bfill()
data12 = pd.read_csv("PRSA_Data_Wanshouxigong_20130301-20170228.csv")
data12 = data12.ffill()
data12 = data12.bfill()
```

```python
# Merge all the data together and drop the unnecessary column "No"
data = pd.concat([data1, data2, data3, data4, data5, data6, data7, data8, data9, data10, data11, data12])
data = data.drop(["No"], axis=1)
```

```python
# Use information from Google Maps and from the original article linked to the dataset to deduce the coordinates of the stations
def latitude_station(x):
    if x == "Aotizhongxin":
        return 40.0013
    if x == "Changping":
        return 40.22477
    if x == "Dingling":
        return 40.3038
    if x == "Dongsi":
        return 39.93069
    if x == "Guanyuan":
        return 39.93345
    if x == "Gucheng":
        return 39.93447
    if x == "Huairou":
        return 40.34949
    if x == "Nongzhanguan":
        return 39.94161
    if x == "Shunyi":
        return 40.13957
    if x == "Tiantan":
        return 39.87545
    if x == "Wanliu":
        return 39.97628
    if x == "Wanshouxigong":
        return 39.87899


def longitude_station(x):
    if x == "Aotizhongxin":
        return 116.38452 
    if x == "Changping":
        return 116.23272
    if x == "Dingling":
        return 116.22932
    if x == "Dongsi":
        return 116.42364 
    if x == "Guanyuan":
        return 116.36949
    if x == "Gucheng":
        return 116.18585
    if x == "Huairou":
        return 116.64917
    if x == "Nongzhanguan":
        return 116.46937
    if x == "Shunyi":
        return 116.64527
    if x == "Tiantan":
        return 116.41221
    if x == "Wanliu":
        return 116.2896
    if x == "Wanshouxigong":
        return 116.37544
```

```python
# Use the tables provided by the US AQI index to code functions that compute the AQI
# Recall tha AQI stands for Air Quality Index

list_ab = [50, 100, 150, 200, 300, 500]
list_un = [0, 51, 101, 151, 201, 301]

def pm25_aqi(pm25):
    list1 = [9., 35.4, 55.4, 125.4, 225.4, 325.4]
    list2 = [0., 9.1, 35.5, 55.5, 125.5, 225.5]
    for i in range(0, 6):
        if pm25 <= list1[i]:
            return np.round( ((list_ab[i] - list_un[i])/(list1[i] - list2[i]))*(pm25 - list2[i]) + list_un[i])
            
    return 500


def pm10_aqi(pm10):
    list1 = [54, 154, 254, 354, 424, 604]
    list2 = [0, 55, 155, 255, 355, 425]
    for i in range(0, 6):
        if pm10 <= list1[i]:
            return np.round( ((list_ab[i] - list_un[i])/(list1[i] - list2[i]))*(pm10 - list2[i]) + list_un[i])
            
    return 500


def no2_aqi(no2):
    list1 = [53, 100, 360, 649, 1249, 2049]
    list2 = [0, 54, 101, 361, 650, 1250]
    for i in range(0, 6):
        if no2 <= list1[i]:
            return np.round( ((list_ab[i] - list_un[i])/(list1[i] - list2[i]))*(no2 - list2[i]) + list_un[i])
            
    return 500


def so2_aqi(so2):
    list1 = [35, 75, 185, 304]
    list2 = [0., 36, 76, 186]
    for i in range(0, 6):
        if so2 <= list1[i]:
            return np.round( ((list_ab[i] - list_un[i])/(list1[i] - list2[i]))*(so2 - list2[i]) + list_un[i])
            
    return 200

def so2_24_aqi(so2_24):
    list1 = [604, 1004]
    list2 = [305, 605]
    if so2_24 >= 305:
        for i in range(0, 2):
            if so2 <= list1[i]:
                return np.round( ((list_ab[i+4] - list_un[i+4])/(list1[i] - list2[i]))*(so2_24 - list2[i]) + list_un[i+4])
        return 500
    else:
        return 0
        


def oz8_aqi(oz8):
    list1 = [0.054, 0.070, 0.085, 0.105, 0.200]
    list2 = [0., 0.055, 0.071, 0.086, 0.106]
    for i in range(0, 5):
        if oz8 <= list1[i]:
            return np.round( ((list_ab[i] - list_un[i])/(list1[i] - list2[i]))*(oz8 - list2[i]) + list_un[i])
            
    return 300


def co_aqi(co):
    list1 = [4.4, 9.4, 12.4, 15.4, 30.4, 50.4]
    list2 = [0., 4.5, 9.5, 12.5, 15.5, 30.5]
    for i in range(0, 6):
        if co <= list1[i]:
            return np.round( ((list_ab[i] - list_un[i])/(list1[i] - list2[i]))*(co - list2[i]) + list_un[i])
            
    return 500


def oz1_aqi(oz1):
    list1 = [0.164, 0.204, 0.404, 0.604]
    list2 = [0.125, 0.165, 0.205, 0.405]
    if oz1 >= 0.125: 
        for i in range(0, 4):
            if oz1 <= list1[i]:
                return np.round( ((list_ab[i+2] - list_un[i+2])/(list1[i] - list2[i]))*(oz1 - list2[i]) + list_un[i+2])
        return 500
    else:
        return 0

def pollution(pm25,pm10,co,no2,so2,so2_24,oz8,oz1):
    aqi = max(pm25_aqi(pm25), pm10_aqi(pm10))
    aqi = max(aqi, co_aqi(co))
    aqi = max(aqi, no2_aqi(no2))
    aqi = max(aqi, so2_aqi(so2))
    aqi = max(aqi, so2_24_aqi(so2_24))
    aqi = max(aqi, oz8_aqi(oz8))
    aqi = max(aqi, oz1_aqi(oz1))

    return aqi



score = np.zeros((17532))

# Now, for each of the considered days, compute the corresponding AQI at 12h
for i in range(0,17532):
    day = data[["PM2.5", "PM10", "SO2", "NO2", "CO", "O3"]][24*i:24*(i+1)]
    day1 = day[12:13]
    day8 = day[8:16]
    aux = day.mean().to_list()
    aux8 = day8.mean().to_list()
    aux1 = day1.mean().to_list()


    pm25 = np.trunc(aux[0]*10)/10
    pm10 = np.trunc(aux[1])
    no2 = np.trunc(0.53*aux1[3])
    so2 = np.trunc(0.38*aux1[2])
    so2_24 = np.trunc(0.38*aux[2])
    oz8 = np.trunc(0.51*aux8[5])/1000
    co = np.trunc(0.87*aux8[4]*0.01)/10
    oz1 = np.trunc(0.51*aux1[5])/1000
    score[i] = pollution(pm25,pm10,co,no2,so2,so2_24,oz8,oz1)
    if i%100 == 0:
        print(i)
```

```python
# Add these values to the dataset
print(len(score))
score = np.repeat(score, 24)
print(len(score))
score = score.astype(np.int64)
data.insert(4,"aqi_index_at_12h", score, True)
```

```python
def level(aqi):
    if 0 <= aqi and aqi <= 50:
        return "Good"
    if 51 <= aqi and aqi <= 100:
        return "Moderate"
    if 101 <= aqi and aqi <= 150:
        return "Unhealthy for Sensitive Groups"
    if 151 <= aqi and aqi <= 200:
        return "Unhealthy"
    if 201 <= aqi and aqi <= 300:
        return "Very Unhealthy"
    if 301 <= aqi and aqi <= 500:
        return "Hazardous"

label = [level(aqi) for aqi in data2["aqi_index_at_12h"]]
```
